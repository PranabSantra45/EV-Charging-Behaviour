#### Data Cleaning 

This is to replace spaces to underscores in column names and to drop rows with all NaN values.

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

df["charging_start_time"] = pd.to_datetime(
    df["charging_start_time"], errors="coerce"
)
df["charging_end_time"] = pd.to_datetime(
    df["charging_end_time"], errors="coerce"
)

# Removing rows with missing start or end time
df = df.dropna(
    subset=["charging_start_time", "charging_end_time"]
)

# Removing invalid sessions means those where end time is before start time
df = df[
    df["charging_end_time"] > df["charging_start_time"]
]

# Creating derived time features
df["charging_duration_minutes"] = (
    df["charging_end_time"] - df["charging_start_time"]
).dt.total_seconds() / 60

df["hour_of_day"] = df["charging_start_time"].dt.hour
df["day_of_week"] = df["charging_start_time"].dt.day_name()

# Remove fully empty rows
df = df.dropna(how="all")
```

```python
df.drop(columns=['time_of_day'], inplace=True)
```

```python
def assign_time_of_day(hour):
    if 0 <= hour <= 5:
        return 'Night'
    elif 6 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 17:
        return 'Afternoon'
    else:
        return 'Evening'

df['time_of_day'] = df['hour_of_day'].apply(assign_time_of_day)
```

```python
df.shape
```

Removed all the rows having missing values for both Energy_consumed and charging_rate

```python
indices = df[(df['energy_consumed_(kwh)'].isnull()) & (df['charging_rate_(kw)'].isnull())].index 
print('indices: ', indices)
df.drop(indices, inplace=True)
```

Filling Missing Values by Median

```python
df['energy_consumed_(kwh)'].fillna(df['energy_consumed_(kwh)'].median(), inplace=True)
df['charging_duration_minutes'].fillna(df['charging_duration_minutes'].median(), inplace=True)
df['charging_rate_(kw)'].fillna(df['charging_rate_(kw)'].median(), inplace=True)
df['distance_driven_(since_last_charge)_(km)'].fillna(df['distance_driven_(since_last_charge)_(km)'].median(), inplace=True)
```

Filling null values for Energy Consumed with Estimated Energy Consumed 

```python
df['Estimated Energy Consumed (kWh)'] = (df['charging_duration_minutes'] / 60) * df['charging_rate_(kw)']

df['energy_consumed_(kwh)'] = df['energy_consumed_(kwh)'].fillna(df['Estimated Energy Consumed (kWh)'])

#add column of charging cost per hour
df['Charging Cost Per Hour (USD)'] = df['charging_cost_(usd)'] / df['charging_duration_(hours)']

#drop na values that still exist
df = df.dropna()
df.info()
print(df.head())
```

Temperature Outliers

```python

df['Temperature (°F)'] = (df['temperature_(â°c)'] * 9/5) + 32

temp_outliers = df.loc[df['Temperature (°F)'] > 105, ['Temperature (°F)', 'time_of_day', 'charging_station_location']]
print('temp_outliers: \n', temp_outliers)

#indices where temperature > 105 F
temp_indices = df[df['Temperature (°F)'] > 105].index 
print('temp_indices: ', temp_indices)

df.drop(temp_indices, inplace=True)

df.info()

temp_buckets = [0, 32, 45, 60, 75, 90, 105]
temp_categories = ['Below Freezing', 'Chilly', 'Cool', 'Mild', 'Hot', 'Sweltering']
df['Temperature Category'] = pd.cut(df['Temperature (°F)'], bins=temp_buckets, labels=temp_categories)

df.head()
df.shape
```

Removing rows for consisting Invalid Values of percentage

```python
df = df[
    (df['state_of_charge_(start_%)'].between(0, 100)) &
    (df['state_of_charge_(end_%)'].between(0, 100))
]
```

Outliers Problem by IQR

```python
def iqr_outlier_mask(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (df[col] < lower) | (df[col] > upper)

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

exclude_cols = [
    'State of Charge (Start %)',
    'State of Charge (End %)'
]

outlier_cols = [col for col in numeric_cols if col not in exclude_cols]

outlier_mask = False

for col in outlier_cols:
    outlier_mask |= iqr_outlier_mask(df, col)
```

## Peak Hour

```python

df_peak = df.copy()

# Find peak hour globally
hourly_energy = df_peak.groupby("hour_of_day")["Estimated Energy Consumed (kWh)"].sum()
peak_hour = hourly_energy.idxmax()

print("Peak Hour Identified:", peak_hour)

# Create classification label
df_peak["is_peak_hour"] = df_peak["hour_of_day"].apply(
    lambda x: 1 if x == peak_hour else 0
)
```

```python

# Targets
y_class = df_peak["is_peak_hour"]
y_reg = df_peak["Estimated Energy Consumed (kWh)"]

# Features (copy again to stay safe)
X_peak = df_peak.drop([
    "Estimated Energy Consumed (kWh)",
    "is_peak_hour"
], axis=1)

# Drop only problematic columns (same as before)
drop_cols = [
    "charging_station_id",
    "charging_station_location",
    "charging_start_time",
    "charging_end_time"
]

for col in drop_cols:
    if col in X_peak.columns:
        X_peak = X_peak.drop(col, axis=1)

# Encoding
X_peak = pd.get_dummies(X_peak, drop_first=True)

print("Feature shape after encoding:", X_peak.shape)
```

Train-Test Split

```python

from sklearn.model_selection import train_test_split

X_train_p, X_test_p, y_class_train, y_class_test = train_test_split(
    X_peak, y_class, test_size=0.2, random_state=42
)

_, _, y_reg_train, y_reg_test = train_test_split(
    X_peak, y_reg, test_size=0.2, random_state=42
)
```

Peak hour (Classification)

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_p, y_class_train)

y_class_pred = clf.predict(X_test_p)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

ConfusionMatrixDisplay.from_predictions(y_class_test, y_class_pred)
plt.show()
```

Energy Prediction (Regression)

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(random_state=42)
reg.fit(X_train_p, y_reg_train)

y_reg_pred = reg.predict(X_test_p)
```

```python
from sklearn.metrics import accuracy_score, mean_absolute_error

print("Peak Hour Accuracy:", accuracy_score(y_class_test, y_class_pred))
print("Energy Prediction MAE:", mean_absolute_error(y_reg_test, y_reg_pred))
```

## Energy Consumption Pipeline

```python
import pandas as pd
import pickle

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor


df = pd.read_csv("Datasets/preprocessed/EV_preprocessed.csv")

df.rename(columns={
    "temperature_(â°c)": "temperature_c"
}, inplace=True)

df["energy_expected"] = (
    (df["state_of_charge_(end_%)"] - df["state_of_charge_(start_%)"]) / 100
) * df["battery_capacity_(kwh)"]


X = df[[
    "hour_of_day",
    "day_of_week",
    "charging_duration_minutes",
    "battery_capacity_(kwh)",
    "state_of_charge_(start_%)",
    "state_of_charge_(end_%)",
    "charging_rate_(kw)",
    "temperature_c",
    "charger_type",
    "location_type",
    "vehicle_model",
    "energy_expected"
]]

y = df["Estimated Energy Consumed (kWh)"]

categorical_cols = [
    "day_of_week",
    "charger_type",
    "location_type",
    "vehicle_model"
]

numeric_cols = [
    "hour_of_day",
    "charging_duration_minutes",
    "battery_capacity_(kwh)",
    "state_of_charge_(start_%)",
    "state_of_charge_(end_%)",
    "charging_rate_(kw)",
    "temperature_c",
    "energy_expected"
]

preprocessor = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
    ("num", "passthrough", numeric_cols)
])

# -------------------------------
# Model Pipeline
# -------------------------------
model = Pipeline([
    ("preprocessing", preprocessor),
    ("regressor", RandomForestRegressor(
        n_estimators=150,
        random_state=42,
        n_jobs=-1
    ))
])

model.fit(X, y)

with open("energy_model_v2.pkl", "wb") as f:
    pickle.dump(model, f)

print(" Model trained and saved as energy_model_v2.pkl")
```
